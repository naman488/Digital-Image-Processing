```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

## Erosion

```python
img = cv2.imread('mor5.png',0)

kernel = np.ones((5,5),np.uint8)

img_erosion_1 = cv2.erode(img,kernel,iterations = 4)
img_erosion_2 = cv2.erode(img,kernel,iterations = 8)

plt.figure(figsize=(16,6))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(img_erosion_1, cmap='gray')
plt.title('Erosion')

plt.subplot(1, 3, 3)
plt.imshow(img_erosion_2, cmap='gray')
plt.title('Erosion');
```

## Dilation

```python
img = cv2.imread('mor5.png',0)

kernel = np.ones((5,5),np.uint8)

img_dilate_1 = cv2.dilate(img,kernel,iterations = 3)
img_dilate_2 = cv2.dilate(img,kernel,iterations = 6)

plt.figure(figsize=(16,6))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(img_dilate_1, cmap='gray')
plt.title('Dilate')

plt.subplot(1, 3, 3)
plt.imshow(img_dilate_2, cmap='gray')
plt.title('Dilate');
```

## Opening

Opening is erosion followed by dilation. Useful in removing background noise

```python
# black image
img =np.zeros((600,600))

# writing text onto the above form image
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,text='ABCDE',org=(50,350), fontFace=font, fontScale= 5, color=(255,255,255),
            thickness=25,lineType=cv2.LINE_AA);
```

```python
plt.figure(figsize=(16,6))
plt.imshow(img, cmap = 'gray');
```

```python
white_noise = np.random.randint(low=0,high=2,size=(600,600))
white_noise = white_noise*255

# adding noise to new formed image
img_noise = white_noise+img

kernel = np.ones((5,5),np.uint8)
img_opening = cv2.morphologyEx(img_noise, cv2.MORPH_OPEN, kernel)

plt.figure(figsize=(16,6))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap = 'gray')
plt.title('Original image')

plt.subplot(1, 3, 2)
plt.imshow(img_noise, cmap = 'gray')
plt.title('Noise image (background)')

plt.subplot(1, 3, 3)
plt.imshow(img_opening, cmap = 'gray')
plt.title('Opening');
```

## Closing

Useful in removing noise from foreground objects, such as black dots on top of the white text.

```python
black_noise = np.random.randint(low=0,high=2,size=(600,600))
black_noise= black_noise * -255

img_black_noise = img + black_noise
img_black_noise[img_black_noise == -255] = 0

kernel = np.ones((5,5),np.uint8)
img_closing = cv2.morphologyEx(img_black_noise, cv2.MORPH_CLOSE, kernel)

plt.figure(figsize=(16,6))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap = 'gray')
plt.title('Original image')

plt.subplot(1, 3, 2)
plt.imshow(img_black_noise, cmap = 'gray')
plt.title('Noise image (foreground)')

plt.subplot(1, 3, 3)
plt.imshow(img_opening, cmap = 'gray')
plt.title('Closing');
```

## Morphological Gradient

Difference between dilation and erosion of an image.

```python
kernel = np.ones((5,5),np.uint8)
img_gradient = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)

plt.figure(figsize=(12,6))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap = 'gray')
plt.title('Original image')

plt.subplot(1, 2, 2)
plt.imshow(img_gradient, cmap = 'gray')
plt.title('Gradient');
```
