```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

### Creating Images

```python
# Making a sqare
square = np.zeros((300, 300), np.uint8)
cv2.rectangle(square, (50, 50), (250, 250), 255, -2)

# Making a ellipse
ellipse = np.zeros((300, 300), np.uint8)
cv2.ellipse(ellipse, (150, 150), (150, 150), 30, 0, 180, 255, -1)

plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.imshow(square, cmap='gray')
plt.title('Square')

plt.subplot(1,2,2)
plt.imshow(ellipse, cmap='gray')
plt.title('Ellipse');
```

## Bitwise operations

### AND operation

```python
# shows only where images intersect
and_op = cv2.bitwise_and(square, ellipse)

plt.figure(figsize=(16,16))

plt.subplot(1,3,1)
plt.imshow(square, cmap='gray')
plt.title('Square')

plt.subplot(1,3,2)
plt.imshow(ellipse, cmap='gray')
plt.title('Ellipse');

plt.subplot(1,3,3)
plt.imshow(and_op, cmap='gray')
plt.title('AND operation');
```

### OR operation

```python
# shows where either square or ellipse is present
or_op = cv2.bitwise_or(square, ellipse)

plt.figure(figsize=(16,16))

plt.subplot(1,3,1)
plt.imshow(square, cmap='gray')
plt.title('Square')

plt.subplot(1,3,2)
plt.imshow(ellipse, cmap='gray')
plt.title('Ellipse');

plt.subplot(1,3,3)
plt.imshow(or_op, cmap='gray')
plt.title('OR operation');
```

### XOR opeartion

```python
# shows where square and ellipse exist by itself
xor_op = cv2.bitwise_xor(square, ellipse)

plt.figure(figsize=(16,16))

plt.subplot(1,3,1)
plt.imshow(square, cmap='gray')
plt.title('Square')

plt.subplot(1,3,2)
plt.imshow(ellipse, cmap='gray')
plt.title('Ellipse');

plt.subplot(1,3,3)
plt.imshow(xor_op, cmap='gray')
plt.title('XOR operation');
```

### NOT operation

```python
# it inverts the image
not_sq_op = cv2.bitwise_not(square)

plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.imshow(square, cmap='gray')
plt.title('Square')

plt.subplot(1,2,2)
plt.imshow(not_sq_op, cmap='gray')
plt.title('NOT - SQUARE');
```

```python
not_sq_op = cv2.bitwise_not(ellipse)

plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.imshow(ellipse, cmap='gray')
plt.title('Ellipse')

plt.subplot(1,2,2)
plt.imshow(not_sq_op, cmap='gray')
plt.title('NOT - ELLIPSE');
```
